```python
import pandas as pd
```

```python
import os
file_path = '../temp/'
files = os.listdir(file_path)
total_df=pd.DataFrame()
for file in files:
    temp= pd.read_csv(f'../temp/{file}')
    total_df=pd.concat([total_df, temp], axis=0)
total_df
```

```python
df = total_df.reset_index(drop=True)
```

```python
df=df.drop('Unnamed: 0', axis=1)
```

```python
df.to_csv('../../daum_economy.csv', index=False)
```
