```python
import pandas as pd
```

```python
df = pd.read_csv('../../daum_economy.csv')
df.dropna(axis=0, subset=['title'])
df.dropna(axis=0, subset=['content'])
df = df[df['content'].apply(lambda x: len(str(x))>50)]

word_list = ['부고', '단신']
text = '|'.join(word_list)
df=df[~df['title'].str.contains(text)]
```

```python
df= df.reset_index(drop=True)
```
